## Data Visualization - Graphical Showing data and patterns fromm the dataset 

```python
import pandas as pd 
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
#  import the dataset 
df = pd.read_csv("dataset.csv")
df.head()
```

## Problem Statement 17 --> 
Visualize the Heart disease dataset by plotting the following graphs using Python. (Define 
objective for every graph)  
a. Histograms    
b. Dot Plots   
c. Bar Plots   
d. Line Charts   
e. Add Histogram and Scatter plot to box plot.

```python
# a) Histogram 
plt.figure( figsize =( 8 , 5 ) , dpi=100)
plt.hist( df['age'] , bins = 20 , edgecolor="black" , color="orange")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of the Age")
plt.show()
```

```python
# b) Dot Plots 
plt.figure( figsize=(8 , 5))
plt.plot(df['age'] , '.' , color= 'blue')
plt.xlabel("Index")
plt.ylabel("age")
plt.title( "Dot Plots of the Age ")
plt.tight_layout()
plt.show()
```

```python
#c) Bar Plot 
plt.figure( figsize=( 8, 5))
count = df['target'].value_counts()
plt.bar( ["No Disease" , "Disease"] ,count ,color='red')
plt.xlabel("Condition")
plt.ylabel("count")
plt.title("Bar Plot Heart Diesease count")
plt.grid()
plt.show()
```

```python
# d) line Chart ( displaying the comparizon of age chlorostle )
plt.figure( figsize=( 8 , 5 ))
sorted_df= df.sort_values( by ='age')
plt.plot( sorted_df['age'], sorted_df['chol'] , color='red')
plt.title("Age v/s Cholorestole")
plt.xlabel("Age")
plt.ylabel("cholestrole")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# E) Adding histgram and Scatter Plot to the bo plot 
# Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(y=df['chol'])

plt.title("Boxplot of Cholesterol")
plt.grid(True)
plt.tight_layout()
plt.show()

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df['chol'], bins=20, edgecolor='black')

plt.title("Histogram of Cholesterol")
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['chol'])

plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.tight_layout()
plt.show()
```
