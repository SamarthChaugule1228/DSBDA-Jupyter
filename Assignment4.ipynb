```python
import pandas as pd
```

```python
# load the Dataset 
df =pd.read_csv("dataset_Facebook.csv")
```

```python
df.head()
```

```python
# Basics of the Info about the Visulaitaion 
print( df.columns)
print(df.info())
print( df.describe())
```

```python
# 3) Data Subsets 
# Columns Subset 
subset = df[['user_id' , 'user_name']]
print(subset.head())
```

```python
# row Filter 
high_likes = df[df['num_replies'] > 18]
print(high_likes.head)
```

```python
# loc(label based )--> give the data from the 3rd index to the 5 th index 
df.loc[3:5,['num_replies' , 'user_id']]
```

```python
# iloc = integer location based indexing

#  It selects data using row index numbers and column index numbers
# 1️⃣ 0:5 → Rows
# Start from row index 0
# Go up to (but not include) 5

# 👉 Selected rows:

# 0, 1, 2, 3, 4   ✅ (5 rows total)
# 2️⃣ 0:2 → Columns
# Start from column index 0
# Go up to (but not include) 2

# 👉 Selected columns:

# 0, 1   ✅ (2 columns total)
print(df.iloc[0:3 , 0 :3])
```

```python
#Sorting the Data 
sorted_df = df.sort_values(by='num_replies' , ascending= False)
print(sorted_df.head(5))
```

```python
# Transpose the Data 
transpose_df = df.T
print(transpose_df)
```

```python
#Melt --> Convert columns into rows 
# wide Format --> many Columns 
# Long Format ---> Fewer Columns , more rows 
df.columns
df = df.dropna(subset=['num_likes','num_replies'])
df['num_likes'] = df['num_likes'].fillna(0)
df['num_replies'] = df['num_replies'].fillna(0)
melt_df = pd.melt(df, id_vars=['type'], value_vars=['num_likes', 'num_replies'])
print(melt_df.head())
```

```python
#Pivot = convert rows into columns 
# long Format ---> many rows 
# wide  Format --> more Columsn 
pivot_df = pd.pivot_table(melt_df, index='type', columns='variable', values='value')
print(pivot_df)
```

```python
# Merge Example 
df2=df[ ['num_replies' , 'num_likes']]
merged = pd.merge( df , df2 , on='num_replies')
print(merged.head())
```
